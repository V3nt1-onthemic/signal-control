### THEORY

#### 1. Lowpass Filter

##### a) Continuous-Time Lowpass Filter

The frequency response $H(j\omega)$ of an ideal continuous-time lowpass filter is defined as follows:

$$ H(j\omega) = \begin{cases} 1, & |\omega| \leq \omega_c \\ 0, & |\omega| > \omega_c \end{cases} \qquad (1) $$

where $\omega_c$ is the cut-off frequency.

For a non-ideal continuous-time lowpass filter, such as a first-order RC lowpass filter, its frequency response $H(j\omega)$ is:

$$ H(j\omega) = \frac{1}{1 + j\omega RC} \qquad (2) $$

where $R$ is the resistance, $C$ is the capacitance. The magnitude and phase of this frequency response are:

$$ |H(j\omega)| = \frac{1}{\sqrt{1 + (\omega RC)^2}} $$

$$ \angle H(j\omega) = -\arctan(\omega RC) $$

##### b) Discrete-Time Lowpass Filter

The frequency response $H(e^{j\omega})$ of an ideal discrete-time lowpass filter is:

$$ H(e^{j\omega}) = \begin{cases} 1, & |\omega| \leq \omega_c \\ 0, & \omega_c < |\omega| \leq \pi \end{cases} \qquad (3) $$

where $\omega_c$ is the cut-off frequency.

---

#### 2. Highpass Filter

##### a) Continuous-Time Highpass Filter

The frequency response $H(j\omega)$ of an ideal continuous-time highpass filter is:

$$ H(j\omega) = \begin{cases} 0, & |\omega| \leq \omega_c \\ 1, & |\omega| > \omega_c \end{cases} \qquad (4) $$

where $\omega_c$ is the cut-off frequency.

The frequency response $H(j\omega)$ of a non-ideal first-order RC highpass filter is:

$$ H(j\omega) = \frac{j\omega RC}{1 + j\omega RC} \qquad (5) $$

The magnitude and phase of this frequency response are:

$$ |H(j\omega)| = \frac{\omega RC}{\sqrt{1 + (\omega RC)^2}} $$

$$ \angle H(j\omega) = \frac{\pi}{2} - \arctan(\omega RC) $$

##### b) Discrete-Time Highpass Filter

The frequency response $H(e^{j\omega})$ of an ideal discrete-time highpass filter is:

$$ H(e^{j\omega}) = \begin{cases} 0, & |\omega| \leq \omega_c \\ 1, & \omega_c < |\omega| \leq \pi \end{cases} \qquad (6) $$

where $\omega_c$ is the cut-off frequency.

---

#### 3. Bandpass Filter

##### a) Continuous-Time Bandpass Filter

The frequency response $H(j\omega)$ of an ideal continuous-time bandpass filter is:

$$ H(j\omega) = \begin{cases} 1, & \omega_1 \leq |\omega| \leq \omega_2 \\ 0, & \text{otherwise} \end{cases} \qquad (7) $$

where $\omega_1$ and $\omega_2$ are the lower and upper limits of the passband frequency, respectively $(\omega_1 < \omega_2)$.

A simple bandpass filter can be obtained by cascading a lowpass filter and a highpass filter. For example, the signal is first passed through a lowpass filter with a cut-off frequency of $\omega_2$, and then passed through a highpass filter with a cut-off frequency of $\omega_1$.

##### b) Discrete-Time Bandpass Filter

The frequency response $H(e^{j\omega})$ of an ideal discrete-time bandpass filter is:

$$ H(e^{j\omega}) = \begin{cases} 1, & \omega_1 \leq |\omega| \leq \omega_2 \\ 0, & \text{otherwise} \end{cases} \qquad (8) $$

where $\omega_1$ and $\omega_2$ satisfy $0 < \omega_1 < \omega_2 < \pi$.

---

#### 4. Finite Impulse Response (FIR) Filter

The FIR filter is a kind of discrete-time filter with a finite-length impulse response. It is also called non-recursive discrete-time filters (discussed in Chapter 3), and the output signal $y[n]$ of the FIR filter is computed as a weighted sum of current and past input signal $x[n]$:

$$ y[n] = \sum_{k = 0}^{N - 1} b_k x[n - k] \qquad (9) $$

where $N$ is the order of the FIR filter, $b_k$ is the coefficients defining the filter's frequency response.

```python
# ============================================================
# Experiment: FIR Filter Design and Signal Filtering
# Pure NumPy version (no scipy required)
# ============================================================

import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# FIR filter design using window method (Hamming window)
# ============================================================

def firwin(numtaps, cutoff, window='hamming', pass_zero='lowpass', fs=1):
    """
    Simple FIR filter design using window method.
    Equivalent to scipy.signal.firwin for basic cases.
    """
    if isinstance(cutoff, (int, float)):
        cutoff = np.atleast_1d(cutoff)
    
    if pass_zero == 'lowpass':
        if len(cutoff) == 1:
            return _fir_lowpass(numtaps, cutoff[0], window, fs)
        else:
            return _fir_bandpass(numtaps, cutoff[0], cutoff[1], window, fs)
    elif pass_zero == 'highpass':
        return _fir_highpass(numtaps, cutoff[0], window, fs)
    elif pass_zero == 'bandpass':
        return _fir_bandpass(numtaps, cutoff[0], cutoff[1], window, fs)
    else:
        raise ValueError("Unsupported pass_zero type")


def _fir_lowpass(numtaps, cutoff, window, fs):
    """Design lowpass FIR filter."""
    n = np.arange(numtaps)
    alpha = (numtaps - 1) / 2
    h = np.sin(2 * np.pi * cutoff / fs * (n - alpha)) / (np.pi * (n - alpha))
    h[int(alpha)] = 2 * cutoff / fs
    return h * _hamming_window(numtaps)


def _fir_highpass(numtaps, cutoff, window, fs):
    """Design highpass FIR filter."""
    n = np.arange(numtaps)
    alpha = (numtaps - 1) / 2
    h = np.sin(2 * np.pi * cutoff / fs * (n - alpha)) / (np.pi * (n - alpha))
    h[int(alpha)] = 2 * cutoff / fs
    h = -h
    h[int(alpha)] += 1
    return h * _hamming_window(numtaps)


def _fir_bandpass(numtaps, lowcut, highcut, window, fs):
    """Design bandpass FIR filter."""
    n = np.arange(numtaps)
    alpha = (numtaps - 1) / 2
    h_lp_high = np.sin(2 * np.pi * highcut / fs * (n - alpha)) / (np.pi * (n - alpha))
    h_lp_high[int(alpha)] = 2 * highcut / fs
    
    h_lp_low = np.sin(2 * np.pi * lowcut / fs * (n - alpha)) / (np.pi * (n - alpha))
    h_lp_low[int(alpha)] = 2 * lowcut / fs
    
    h = h_lp_high - h_lp_low
    return h * _hamming_window(numtaps)


def _hamming_window(numtaps):
    """Generate Hamming window."""
    n = np.arange(numtaps)
    return 0.54 - 0.46 * np.cos(2 * np.pi * n / (numtaps - 1))


def freqz(b, a=1, worN=512, fs=1):
    """
    Simple frequency response calculation.
    Equivalent to scipy.signal.freqz.
    """
    # Convert a to array if it's a scalar
    if np.isscalar(a):
        a = np.array([a])
    else:
        a = np.asarray(a)
    
    w = np.linspace(0, np.pi, worN)
    h = np.zeros(w.shape, dtype=complex)
    for i, freq in enumerate(w):
        z = np.exp(1j * freq)
        h_num = np.sum(b * (z ** -np.arange(len(b))))
        h_den = np.sum(a * (z ** -np.arange(len(a))))
        h[i] = h_num / h_den
    return w, h


# ============================================================
# Generate test signal
# ============================================================

fs = 1000  # Sampling frequency (Hz)
t = np.arange(0, 1, 1/fs)  # Time vector (1 second)

# Signal containing three frequency components
x = np.sin(2 * np.pi * 50 * t) + np.sin(2 * np.pi * 200 * t) + np.sin(2 * np.pi * 400 * t)

# Nyquist frequency
nyquist = fs / 2  # 500 Hz

# ============================================================
# Design FIR filters using custom firwin
# ============================================================

# Filter order
N = 60

# (1) Lowpass filter: cut-off at 100 Hz
fc_lp = 100  # Hz
wn_lp = fc_lp / nyquist  # Normalized cut-off frequency
b_lp = firwin(N + 1, wn_lp, window='hamming', pass_zero='lowpass', fs=1)

# (2) Highpass filter: cut-off at 300 Hz
fc_hp = 300  # Hz
wn_hp = fc_hp / nyquist  # Normalized cut-off frequency
b_hp = firwin(N + 1, wn_hp, window='hamming', pass_zero='highpass', fs=1)

# (3) Bandpass filter: passband 150-250 Hz
f1_bp = 150  # Hz
f2_bp = 250  # Hz
wn_bp = [f1_bp / nyquist, f2_bp / nyquist]  # Normalized passband
b_bp = firwin(N + 1, wn_bp, window='hamming', pass_zero='bandpass', fs=1)

# ============================================================
# Apply filters using convolution
# ============================================================

y_lp = np.convolve(x, b_lp, mode='same')
y_hp = np.convolve(x, b_hp, mode='same')
y_bp = np.convolve(x, b_bp, mode='same')

# ============================================================
# Frequency response of filters
# ============================================================

w_lp, h_lp = freqz(b_lp, 1, worN=2000, fs=1)
w_hp, h_hp = freqz(b_hp, 1, worN=2000, fs=1)
w_bp, h_bp = freqz(b_bp, 1, worN=2000, fs=1)

# Convert from normalized frequency (0 to pi) to Hz
freq_hz_lp = w_lp * nyquist / np.pi
freq_hz_hp = w_hp * nyquist / np.pi
freq_hz_bp = w_bp * nyquist / np.pi

# ============================================================
# Plot 1: Original signal and filtered signals (time domain)
# ============================================================

fig, axes = plt.subplots(4, 1, figsize=(12, 10))

# Plot time range: 0 to 0.1s for better visualization
t_zoom = 0.1
idx_zoom = int(t_zoom * fs)

axes[0].plot(t[:idx_zoom], x[:idx_zoom], 'b-', linewidth=1.5)
axes[0].set_title('Original Signal (Time Domain)')
axes[0].set_xlabel('Time (s)')
axes[0].set_ylabel('Amplitude')
axes[0].grid(True)

axes[1].plot(t[:idx_zoom], y_lp[:idx_zoom], 'r-', linewidth=1.5)
axes[1].set_title('Lowpass Filtered Signal (fc = 100 Hz)')
axes[1].set_xlabel('Time (s)')
axes[1].set_ylabel('Amplitude')
axes[1].grid(True)

axes[2].plot(t[:idx_zoom], y_hp[:idx_zoom], 'g-', linewidth=1.5)
axes[2].set_title('Highpass Filtered Signal (fc = 300 Hz)')
axes[2].set_xlabel('Time (s)')
axes[2].set_ylabel('Amplitude')
axes[2].grid(True)

axes[3].plot(t[:idx_zoom], y_bp[:idx_zoom], 'm-', linewidth=1.5)
axes[3].set_title('Bandpass Filtered Signal (150-250 Hz)')
axes[3].set_xlabel('Time (s)')
axes[3].set_ylabel('Amplitude')
axes[3].grid(True)

plt.tight_layout()
plt.savefig('time_domain_signals.png', dpi=150)
plt.show()

# ============================================================
# Plot 2: Frequency spectrum (FFT)
# ============================================================

# Compute FFT
X = np.fft.fft(x)
Y_lp = np.fft.fft(y_lp)
Y_hp = np.fft.fft(y_hp)
Y_bp = np.fft.fft(y_bp)

freq_axis = np.fft.fftfreq(len(x), 1/fs)

# Only plot positive frequencies up to Nyquist
idx_pos = freq_axis >= 0

fig, axes = plt.subplots(4, 1, figsize=(12, 10))

axes[0].plot(freq_axis[idx_pos], np.abs(X[idx_pos]), 'b-', linewidth=1.5)
axes[0].set_title('Original Signal (Frequency Domain)')
axes[0].set_xlabel('Frequency (Hz)')
axes[0].set_ylabel('Magnitude')
axes[0].grid(True)
axes[0].set_xlim([0, 500])

axes[1].plot(freq_axis[idx_pos], np.abs(Y_lp[idx_pos]), 'r-', linewidth=1.5)
axes[1].set_title('Lowpass Filtered Signal (fc = 100 Hz)')
axes[1].set_xlabel('Frequency (Hz)')
axes[1].set_ylabel('Magnitude')
axes[1].grid(True)
axes[1].set_xlim([0, 500])

axes[2].plot(freq_axis[idx_pos], np.abs(Y_hp[idx_pos]), 'g-', linewidth=1.5)
axes[2].set_title('Highpass Filtered Signal (fc = 300 Hz)')
axes[2].set_xlabel('Frequency (Hz)')
axes[2].set_ylabel('Magnitude')
axes[2].grid(True)
axes[2].set_xlim([0, 500])

axes[3].plot(freq_axis[idx_pos], np.abs(Y_bp[idx_pos]), 'm-', linewidth=1.5)
axes[3].set_title('Bandpass Filtered Signal (150-250 Hz)')
axes[3].set_xlabel('Frequency (Hz)')
axes[3].set_ylabel('Magnitude')
axes[3].grid(True)
axes[3].set_xlim([0, 500])

plt.tight_layout()
plt.savefig('frequency_spectrum.png', dpi=150)
plt.show()

# ============================================================
# Plot 3: Filter frequency response (Magnitude)
# ============================================================

fig, axes = plt.subplots(3, 1, figsize=(12, 10))

axes[0].plot(freq_hz_lp, 20 * np.log10(np.abs(h_lp) + 1e-10), 'b-', linewidth=1.5)
axes[0].axvline(x=100, color='r', linestyle='--', linewidth=1, label='Cut-off: 100 Hz')
axes[0].set_title('Lowpass Filter Frequency Response (Order = 60)')
axes[0].set_xlabel('Frequency (Hz)')
axes[0].set_ylabel('Magnitude (dB)')
axes[0].set_xlim([0, 500])
axes[0].set_ylim([-80, 10])
axes[0].grid(True)
axes[0].legend()

axes[1].plot(freq_hz_hp, 20 * np.log10(np.abs(h_hp) + 1e-10), 'g-', linewidth=1.5)
axes[1].axvline(x=300, color='r', linestyle='--', linewidth=1, label='Cut-off: 300 Hz')
axes[1].set_title('Highpass Filter Frequency Response (Order = 60)')
axes[1].set_xlabel('Frequency (Hz)')
axes[1].set_ylabel('Magnitude (dB)')
axes[1].set_xlim([0, 500])
axes[1].set_ylim([-80, 10])
axes[1].grid(True)
axes[1].legend()

axes[2].plot(freq_hz_bp, 20 * np.log10(np.abs(h_bp) + 1e-10), 'm-', linewidth=1.5)
axes[2].axvline(x=150, color='r', linestyle='--', linewidth=1, label='Passband: 150-250 Hz')
axes[2].axvline(x=250, color='r', linestyle='--', linewidth=1)
axes[2].set_title('Bandpass Filter Frequency Response (Order = 60)')
axes[2].set_xlabel('Frequency (Hz)')
axes[2].set_ylabel('Magnitude (dB)')
axes[2].set_xlim([0, 500])
axes[2].set_ylim([-80, 10])
axes[2].grid(True)
axes[2].legend()

plt.tight_layout()
plt.savefig('filter_frequency_response.png', dpi=150)
plt.show()

# ============================================================
# Plot 4: Impulse responses of filters
# ============================================================

fig, axes = plt.subplots(3, 1, figsize=(12, 8))

n_impulse = np.arange(len(b_lp))
axes[0].stem(n_impulse, b_lp, basefmt=' ')
axes[0].set_title('Lowpass Filter Impulse Response (Order = 60)')
axes[0].set_xlabel('Sample Index')
axes[0].set_ylabel('Amplitude')
axes[0].grid(True)

axes[1].stem(n_impulse, b_hp, basefmt=' ')
axes[1].set_title('Highpass Filter Impulse Response (Order = 60)')
axes[1].set_xlabel('Sample Index')
axes[1].set_ylabel('Amplitude')
axes[1].grid(True)

axes[2].stem(n_impulse, b_bp, basefmt=' ')
axes[2].set_title('Bandpass Filter Impulse Response (Order = 60)')
axes[2].set_xlabel('Sample Index')
axes[2].set_ylabel('Amplitude')
axes[2].grid(True)

plt.tight_layout()
plt.savefig('impulse_responses.png', dpi=150)
plt.show()

# ============================================================
# Print summary
# ============================================================

print("=" * 60)
print("Filtering Experiment Results Summary")
print("=" * 60)

print("\nInput Signal Components:")
print("  50 Hz (sinusoid)")
print("  200 Hz (sinusoid)")
print("  400 Hz (sinusoid)")

print("\nFilter Specifications:")
print(f"  Filter Order: {N}")
print(f"  Nyquist Frequency: {nyquist} Hz")
print(f"  Lowpass: cut-off = {fc_lp} Hz (normalized: {wn_lp:.3f})")
print(f"  Highpass: cut-off = {fc_hp} Hz (normalized: {wn_hp:.3f})")
print(f"  Bandpass: passband = {f1_bp}-{f2_bp} Hz (normalized: {wn_bp[0]:.3f}-{wn_bp[1]:.3f})")

print("\nFiltering Results:")
print("  Lowpass Filter: 50 Hz component passes, 200 Hz and 400 Hz are attenuated")
print("  Highpass Filter: 400 Hz component passes, 50 Hz and 200 Hz are attenuated")
print("  Bandpass Filter: 200 Hz component passes, 50 Hz and 400 Hz are attenuated")

print("\n" + "=" * 60)
print("All plots have been saved successfully!")
print("=" * 60)

# ============================================================
# Additional: Compare effect of different filter orders
# ============================================================

orders = [10, 30, 60]
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

for idx, order in enumerate(orders):
    b_test = firwin(order + 1, wn_lp, window='hamming', pass_zero='lowpass', fs=1)
    w_test, h_test = freqz(b_test, 1, worN=2000, fs=1)
    freq_test = w_test * nyquist / np.pi
    
    axes[idx].plot(freq_test, 20 * np.log10(np.abs(h_test) + 1e-10), 'b-', linewidth=1.5)
    axes[idx].axvline(x=100, color='r', linestyle='--', linewidth=1, label='Cut-off: 100 Hz')
    axes[idx].set_title(f'Lowpass Filter (Order = {order})')
    axes[idx].set_xlabel('Frequency (Hz)')
    axes[idx].set_ylabel('Magnitude (dB)')
    axes[idx].set_xlim([0, 500])
    axes[idx].set_ylim([-80, 10])
    axes[idx].grid(True)
    axes[idx].legend()

plt.tight_layout()
plt.savefig('filter_order_comparison.png', dpi=150)
plt.show()

print("\nAdditional plot saved: filter_order_comparison.png")
print("Higher order filters provide steeper roll-off but require more computation.")
```

### RELATED QUESTIONS:

#### 1. In this experiment, we used `fir1` to design FIR lowpass, highpass, and bandpass filters and applied convolution for signal processing. Why do FIR filters require a higher filter order to achieve steeper filter characteristics?

**Answer:**

FIR filters require a higher filter order to achieve steeper filter characteristics because of the following reasons:

**Theoretical Explanation:**

The frequency response of an FIR filter is determined by the length of its impulse response. The transition band width (the frequency range between the passband and stopband) is inversely proportional to the filter order:

$$ \text{Transition Width} \propto \frac{1}{N} $$

where $N$ is the filter order. Therefore, to achieve a narrower transition band (steeper roll-off), a larger $N$ is required.

**Practical Implications:**

| Filter Order | Transition Band Width | Filter Characteristics |
|--------------|----------------------|----------------------|
| Low ($N$ small) | Wide | Slow roll-off, poor frequency selectivity |
| Medium ($N$ moderate) | Moderate | Moderate roll-off, acceptable selectivity |
| High ($N$ large) | Narrow | Sharp roll-off, excellent frequency selectivity |

**Why FIR filters cannot achieve arbitrarily steep roll-off with low order:**

1. **Finite Impulse Response:** The filter's response is limited to a finite number of samples. Shorter impulse responses cannot accurately approximate the ideal infinite-length impulse response required for perfect brick-wall filtering.

2. **Uncertainty Principle:** There is a fundamental trade-off between time-domain and frequency-domain resolution. A shorter filter (lower order) has better time resolution but poorer frequency resolution.

3. **Windowing Effect:** When designing FIR filters using the window method (as in `fir1`), the filter order determines the length of the window. A longer window produces a frequency response closer to the ideal response.

**Comparison:**

| Filter Type | Typical Order for Sharp Transition |
|-------------|-------------------------------------|
| FIR (Window method) | High (e.g., 50-100+) |
| IIR (Butterworth, Chebyshev) | Low (e.g., 4-8) |

#### 3. In the experiment, both the filter order and the cut-off frequency influence the filtering effect. What problems might arise if the order of the filters is too low?

**Answer:**

If the filter order is too low, the following problems may arise:

**1. Wide Transition Band (Poor Selectivity)**

- The frequency response rolls off slowly, meaning frequencies near the cut-off are not properly attenuated or passed.
- For example, with a low-order lowpass filter designed for 100 Hz, the 200 Hz component may not be fully attenuated.
- This leads to poor frequency separation.

**2. Stopband Attenuation is Insufficient**

- Frequencies that should be completely removed (stopband) may still have significant amplitude.
- This results in residual unwanted frequency components in the output signal.

**3. Passband Ripples**

- Low-order filters may introduce ripples in the passband, causing signal distortion.
- The amplitude of desired frequency components may be affected.

**4. Gibbs Phenomenon (Overshoot)**

- Low-order FIR filters designed using the window method may exhibit significant overshoot near discontinuities (if using abrupt windows).

**5. Inability to Remove Specific Frequency Components**

- With insufficient order, the filter cannot effectively separate closely spaced frequencies.
- For example, separating 200 Hz and 250 Hz requires higher order than separating 50 Hz and 400 Hz.

**6. Example Comparison:**

| Order | Performance |
|-------|-------------|
| N = 10 | Very slow roll-off, poor attenuation of unwanted frequencies, significant residual components |
| N = 30 | Moderate roll-off, acceptable attenuation |
| N = 60 | Sharp roll-off, good attenuation, clean separation of desired frequencies |

**Visual Demonstration:**

In the filter order comparison plots (with orders 10, 30, and 60), you can observe:

- **Order 10:** The stopband attenuation is only -20 dB to -30 dB; unwanted frequencies leak through.
- **Order 30:** Stopband attenuation improves to -40 dB to -50 dB.
- **Order 60:** Stopband attenuation reaches -60 dB to -80 dB, providing excellent frequency selectivity.

**Conclusion:**

Using too low a filter order results in poor filtering performance. However, using an excessively high order also has drawbacks:
- Increased computational cost
- Longer processing time
- Increased memory requirements
- Higher latency (more samples to process)

Therefore, the filter order should be chosen based on the specific application requirements, balancing performance and computational efficiency.
