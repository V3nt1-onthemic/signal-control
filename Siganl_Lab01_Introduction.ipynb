### 冲激函数（Impulse Function）

冲激函数定义如下：

$$\int_{-\infty}^{+\infty}\delta(t)dt = 1$$

且当 $t \neq 0$ 时，$\delta(t) = 0$

即冲激函数在除 $t=0$ 以外的所有点处幅值为零。

### 阶跃函数（Step Function）

单位阶跃函数定义如下：

$$u(t) = \begin{cases} 1, & t \geq 0 \\ 0, & t < 0 \end{cases}$$

### 实验二：二维/三维绘图
#### 1.冲激函数
#### 2.阶跃函数
#### 3.正弦函数
#### 4.三维绘图

```python
import numpy as np
import matplotlib.pyplot as plt

# 时间轴
t = np.arange(-2, 10.01, 0.01)

# 1. 冲激函数
impulse = np.zeros_like(t)
idx = np.argmin(np.abs(t))
impulse[idx] = 1.0

# 2. 阶跃函数
unitstep = (t >= 0).astype(float)

# 3. 斜坡函数
ramp = t * unitstep

# 绘制冲激函数（修正版 - 去掉 use_line_collection）
plt.figure(figsize=(10, 4))
plt.stem(t, impulse, basefmt=' ', linefmt='k-', markerfmt='ko')
plt.ylim([-0.05, 1.05])
plt.title('Impulse Function')
plt.grid(True, alpha=0.3)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# 绘制阶跃函数
plt.figure(figsize=(10, 4))
plt.plot(t, unitstep, 'k-', linewidth=2)
plt.ylim([-0.05, 1.05])
plt.title('Unit Step Function')
plt.grid(True, alpha=0.3)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

# 绘制斜坡函数
plt.figure(figsize=(10, 4))
plt.plot(t, ramp, 'k-', linewidth=2)
plt.ylim([-0.5, 10.5])
plt.title('Ramp Function')
plt.grid(True, alpha=0.3)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

### 问题 1：物理中的阻尼振荡

阻尼振荡出现在机械振动、RLC 电路和量子波函数等物理系统中。振荡行为受阻尼因子影响，代表系统中的能量损失。

信号表达式：
$$x(t) = Ae^{-\alpha t}\sin(\omega t)$$

其中：
- $A$ 为振幅
- $\alpha$ 为阻尼系数（值越大衰减越快）
- $\omega$ 为角频率

#### 任务

1. 在 Python 中定义 $t$，范围 0 到 10，步长 0.01
2. 绘制 $x(t)$ 在不同 $\alpha$ 值（0.1、0.5、1）下的曲线，保持 $A = 1$，$\omega = 2\pi$
3. 比较衰减速率，描述阻尼对振荡的影响

#### 附加挑战 - 能量耗散

计算并绘制包络线 $Ae^{-\alpha t}$，代表能量随时间的损失。增加 $\alpha$ 如何影响系统稳定性？

#### 代码实现


```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
t = np.arange(0, 10.01, 0.01)
A = 1
omega = 2 * np.pi
alpha_values = [0.1, 0.5, 1.0]
colors = ['blue', 'red', 'green']

# 创建图形
plt.figure(figsize=(12, 6))

for i, alpha in enumerate(alpha_values):
    # 计算阻尼振荡信号
    x = A * np.exp(-alpha * t) * np.sin(omega * t)
    
    # 绘制信号
    plt.plot(t, x, color=colors[i], linewidth=1.5, 
             label=f'$\\alpha$ = {alpha}')
    
    # 绘制包络线
    envelope = A * np.exp(-alpha * t)
    plt.plot(t, envelope, '--', color=colors[i], linewidth=1, alpha=0.7)
    plt.plot(t, -envelope, '--', color=colors[i], linewidth=1, alpha=0.7)

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Damped Oscillations with Different Damping Coefficients')
plt.legend(loc='best')
plt.grid(True, alpha=0.3)
plt.xlim([0, 10])
plt.show()
```

#### 子图对比

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 10.01, 0.01)
A = 1
omega = 2 * np.pi
alpha_values = [0.1, 0.5, 1.0]

fig, axes = plt.subplots(3, 1, figsize=(12, 10))

for i, alpha in enumerate(alpha_values):
    x = A * np.exp(-alpha * t) * np.sin(omega * t)
    envelope = A * np.exp(-alpha * t)
    
    axes[i].plot(t, x, 'b-', linewidth=1.5)
    axes[i].plot(t, envelope, 'r--', linewidth=1, alpha=0.7)
    axes[i].plot(t, -envelope, 'r--', linewidth=1, alpha=0.7)
    axes[i].set_title(f'Damping Coefficient $\\alpha$ = {alpha}')
    axes[i].set_xlabel('Time (s)')
    axes[i].set_ylabel('Amplitude')
    axes[i].grid(True, alpha=0.3)
    axes[i].set_xlim([0, 10])

plt.tight_layout()
plt.show()
```

### 问题 2：电磁波偏振矩阵

在电磁波中，偏振描述电场的方向。**琼斯矩阵（Jones Matrix）**用数学方式表示偏振态：

$$E = \begin{bmatrix} E_x \\ E_y \end{bmatrix}$$

$2 \times 2$ 矩阵可描述相移或双折射等变换。该矩阵的**特征值和行列式**揭示重要性质：
- **特征值**对应特征偏振模式
- **行列式**表示变换是否保持信号强度（行列式 = 1）或改变它

#### 任务

1. 定义一个 $2 \times 2$ 复矩阵 $E$ 表示电场变换
2. 使用 `np.linalg.eig(E)` 计算其特征值和特征向量
3. 使用 `np.linalg.det(E)` 计算其行列式

#### 附加挑战 - 奇异与非奇异

改变 $E$ 中的元素，观察 $\det(E) = 0$ 的情况。这对波传输意味着什么？

```python
import numpy as np

# 定义琼斯矩阵
E = np.array([[3+4j, 2-1j], 
              [1+2j, 4+3j]])

print("琼斯矩阵 E:")
print(E)
print()

# 计算特征值和特征向量
eigenvalues, eigenvectors = np.linalg.eig(E)

print("特征值:")
print(eigenvalues)
print()

print("特征向量:")
print(eigenvectors)
print()

# 计算行列式
det_E = np.linalg.det(E)
print(f"行列式 det(E) = {det_E}")
print()

# 验证特征方程 E·v = λ·v
print("验证特征方程 E·v = λ·v:")
for i in range(2):
    v = eigenvectors[:, i]
    lambda_i = eigenvalues[i]
    result = E @ v - lambda_i * v
    print(f"  特征向量 {i+1}: {result}")
```

#### 奇异与非奇异矩阵测试

```python
import numpy as np

matrices = {
    '非奇异矩阵': np.array([[1, 2], [3, 4]]),
    '奇异矩阵': np.array([[1, 2], [2, 4]]),
    '单位矩阵': np.eye(2),
    '零矩阵': np.zeros((2, 2)),
    '琼斯矩阵': np.array([[3+4j, 2-1j], [1+2j, 4+3j]])
}

for name, M in matrices.items():
    det = np.linalg.det(M)
    print(f"{name}:")
    print(M)
    print(f"行列式 = {det:.4f}")
    if abs(det) < 1e-10:
        print("→ 矩阵奇异，波传输可能受阻或消失")
    else:
        print("→ 矩阵非奇异，波可以正常传输")
    print()
```
